```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Guinevere", "Stefanie", "Bernardina", "Khariton", "Kristina", np.nan, "Sampath"],
    "last": ["Kovalyov", "Nagi", "Von Essen", "Valenta", "Albini", "N/A", "Katz"],
    "emails": ['GuinevereKovalyov@example.com', 'StefanieNagi@example.com', 'BernardinaVonEssen@example.com', 
               'KharitonValenta@example.com', 'KristinaAlbini@example.com', 'EleonoraReagan@example.com', "Missing"]
}
```

```python
df = pd.DataFrame(people)

df.replace("N/A", np.nan, inplace=True)
df.replace("Missing", np.nan, inplace=True)

df
```

```python
df.dropna(axis="index", how="any", inplace=True)
```
